```python
X =new_data.drop('Class',axis=1) 
y=new_data['Class']
```

# Updating for real time detection and notify the user about the transcation

```python
import numpy as np
import pickle
import cx_Oracle
import threading
import time
import plyer  

# Load the trained Random Forest model
try:
    with open("credit_card_model", "rb") as file:
        model = pickle.load(file)
except FileNotFoundError:
    model = None


features = ["Time", "V1", "V2", "V3", "V4", "V5", "V6", "V7", "V8", "V9", "V10", "V11", "V12", "V13", "V14", "V15", "V16", "V17", "V18", "V19", "V20", "V21", "V22", "V23", "V24", "V25", "V26", "V27", "V28", "Amount"]


DB_DSN = cx_Oracle.makedsn("localhost", "1521", service_name="orcl")
DB_USER = "my_database"
DB_PASSWORD = "ssah@123"

last_transaction = None  

# Fetch latest transaction from Oracle database
def fetch_latest_transaction():
    try:
        connection = cx_Oracle.connect(DB_USER, DB_PASSWORD, DB_DSN)
        cursor = connection.cursor()
        query = f"""
            SELECT {', '.join(feature_names)}
            FROM transactions
            ORDER BY transaction_time DESC FETCH FIRST 1 ROWS ONLY
        """
        cursor.execute(query)
        row = cursor.fetchone()
        cursor.close()
        connection.close()
        return row if row else None
    except cx_Oracle.DatabaseError as e:
        print(f"Database Error: {e}")
        return None

# Send system notification
def send_notification(title, message):
    plyer.notification.notify(title=title, message=message, timeout=5)

# Predict fraud
def predict_fraud():
    global last_transaction
    if model is None:
        print("Error: Model not found! Please train and save the model first.")
        return
    
    transaction = fetch_latest_transaction()
    if transaction is None:
        print("No new transaction found.")
        return
    
    if transaction == last_transaction:
        return  
    
    last_transaction = transaction  
    input_array = np.array(transaction).reshape(1, -1)
    prediction = model.predict(input_array)[0]
    
    print("Transaction Details:")
    print(dict(zip(feature_names, transaction)))
    
    if prediction == 1:
        message = "ALERT: Fraudulent Transaction Detected!"
        print(message)
        send_notification("Fraud Alert", message)
    else:
        message = "Transaction is Legitimate."
        print(message)
        send_notification("Transaction Status", message)

# Real-time transaction monitoring
def monitor_transactions():
    while True:
        predict_fraud()
        time.sleep(20)  


threading.Thread(target=monitor_transactions, daemon=True).start()


while True:
    time.sleep(1)
```
